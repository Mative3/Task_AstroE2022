En este notebook voy a presentar un problema relacionado con mi trabajo, que espero poder resolver utilizando las técnicas aprendidas en el curso Introducción a la Astroestadística 2022.

En mi trabajo estoy modelando atmósferas de estrellas enanas blancas magnéticas (MWDs por sus siglas en inglés). Estas atmósferas se caracterizan por estar altamente magnetizadas, con campos magnéticos tan intensos como $B=10^6-10^9 \,$ G (para describir la intensidad del campo magnético se suele utilizar un parámetro $\gamma=B/B_0$, con $B_0=4.70103 \times 10^9 \,$ G). Además, el principal constituyente de estas atmósferas es el hidrógeno, con una fuerte preponderancia de este elemento en su forma atómica. Por lo tanto, comprender cómo es afectado un átomo de hidrógeno por un campo magnético súper intenso es crucial para el desarrollo del modelo estelar.

En particular, lo que pretendo analizar en este curso es la variación del tamaño efectivo del átomo de hidrógeno en función de la intensidad del campo magnético, como así también del estado cuántico en el que éste se encuentra. Esto corresponde a un problema de regresión multivariable, en el que no conozco la forma que tendrá la función de ajuste.

Un átomo de hidrógeno en un campo magnético está caracterizado por tres números cuánticos: (N,m,$\nu$), donde N es el número cuántico de Landau, m el número cuántico magnético, y $\nu$ un número cuántico longitudinal. Estos números correlacionan directamente con aquellos que caracterizan al átomo en ausencia de campo. 

En particular, a mí me interesan aquellos estados en que $N=0$, por lo que los únicos números libres son m y $\nu$. 

¿Cómo varían estos números cuánticos?

$m=0, \,-1, \,-2, \,-3, \,-4, \,-5, \,-6, \, \ldots$,

$\nu=0, \, +1, \, +2, \, +3, \, +4, \, +5, \, +6, \, \ldots$.

Además, los datos de los que yo dispongo incluyen sólo aquellos estados tales que $m=0, \, -1, \, -2, \, -3, \,-4$ y $\nu=0,\, +1, \ldots, \, +60$, los cuales son obtenidos con un modelo. Para estos estados conozco, entre otras cosas, el tamaño efectivo en función de $\gamma$. Sin embargo, el ćalculo de cada uno de estos estados es complicado y demanda mucho tiempo de cómputo. Por lo tanto, de los 300 estados a los que tengo acceso, la idea sería trabajar con el menor número de puntos posibles.

Así, lo que yo espero es intentar calcular una función multivariable que ajuste el tamaño efectivo del átomo para cada valor de $(\gamma,m,\nu)$ conocidos, y que describa el comportamiento para todos los estados $(m,\nu)$, cualesquiera que sean estos números. En principio, debería existir algún tipo de convergencia cuando $m \rightarrow -\infty$ y $\nu \rightarrow +\infty$. 

A continuación muestro cómo varían tres de los estados conocidos del átomo cuando éste se ve sometido a la acción de un campo magnético de intensidad arbitraria. Las columnas de la siguiente figura representan los tres estados analizados, que se etiquetan en la cabecera de cada una de ellas. Las filas representan distintas intensidades de campo magnético, que se etiquetan a la izquierda de cada una de ellas mediante valores del parámetro $\gamma$. Las gráficas muestran curvas de nivel de la distribución $\left|\Psi\right|^2$, con $\Psi$ la función de onda, siendo la línea de trazos un nivel de referencia. Estas curvas de nivel se consideran representativas del tamaño efectivo del átomo. Por lo tanto, se puede observar que un campo magnético tiende a comprimir el átomo según su intensidad, a diferentes velocidades en la dirección $z$ (dirección del campo) y en la dirección $x$ (dirección transversal al campo).

```python
import numpy as np
import matplotlib
import math
import matplotlib.pylab as plt
from matplotlib.transforms import offset_copy
from pathlib import Path
from scipy.interpolate import griddata
```

```python
gamma=np.array([1e-5,1e-1,1e0,1e1,1e2]);

fig,ax=plt.subplots(len(gamma),3,figsize=(15,20));

rows = ['{}'.format(col) for col in ['$ \gamma=10^{-5} $', '$\gamma=10^{-1}$', '$\gamma=10^0$', '$\gamma=10^1$', '$\gamma=10^2$']];
cols = ['{}'.format(row) for row in ['$1s_0$', '$2p_0$', '$2p_{-1}$']]

plt.setp(ax.flat, xlabel='$\Delta x/a_{0}$', ylabel='$\Delta z/a_{0}$');

for i in range(len(gamma)):
    
    for j in range(3):
        
        if(j==0):
            
            fname='1H0+1_'+str(format(gamma[i],'.5f'))+'.2dbb.plotdata';
            name=Path('Datos','Reference','1s0',fname);
            
            x0=3;
            y0=3;
            
        elif(j==1):

            fname='1H0-1_'+str(format(gamma[i],'.5f'))+'.2dbb.plotdata';
            name=Path('Datos','Reference','2p0',fname);

            x0=10;
            y0=10;
            
        elif(j==2):
            
            fname='1H1+1_'+str(format(gamma[i],'.5f'))+'.2dbb.plotdata';
            name=Path('Datos','Reference','2p-1',fname);

            x0=10;
            y0=10;


        A=np.loadtxt(name);

        x=A[:,0];
        y=A[:,1];
        z=A[:,2];

        N=len(x);

        contA=-1;
        contB=0;

        X=np.zeros((500,300));
        Y=np.zeros((500,300));
        Z=np.zeros((500,300));

        for k in range(N-1):

            contA=contA+1;

            X[contA,contB]=x[k];
            Y[contA,contB]=y[k];
            Z[contA,contB]=z[k];

            if(abs(x[k+1]-x[k])>0):

                contB=contB+1;
                contA=0;


        if(i==0 and j==2):
            for k in range(500):
                Z[k,29]=(Z[k,28]+Z[k,30])/2;

                
        xvec=np.linspace(x[0],x[N-1],100);
        yvec=np.linspace(y[0],y[N-1],100);

        xx,yy=np.meshgrid(xvec,yvec);

        zz=griddata((X.ravel(),Y.ravel()),Z.ravel(),(xx,yy),method='linear'); 
        
        index=np.argwhere(np.isnan(zz));
        
        zz[index]=-1;
        
        zz_max=np.max(zz);
        
        zz[index]='nan'
        
        zzN=zz/zz_max;

        ax[i,j].contour(xx,yy,zzN,10);
        ax[i,j].contour(xx,yy,zzN,levels=[1/100],colors='k',linestyles='dashed');

        ax[i,j].axis('square')

        ax[i,j].set_xlim([0,x0])
        ax[i,j].set_ylim([0,y0])
        
        pad=5;
        
        for axes,col in zip(ax[0],cols):
            axes.annotate(col,xy=(0.5,1),xytext=(0,pad),xycoords='axes fraction',textcoords='offset points',size='large',ha='center',va='baseline');
        
        for axes,row in zip(ax[:,0],rows):
            axes.annotate(row,xy=(0, 0.5),xytext=(-axes.yaxis.labelpad-pad,0),xycoords=axes.yaxis.label,textcoords='offset points',size='large',ha='right',va='center');
```

La cuestión ahora es cómo puedo reducir el ruido. Veamos qué nos dice la teoría: Cuando $\gamma \rightarrow 0$, el radio efectivo del estado $\left(n,l,m\right)$ está dado por

$$R_{nl}= \frac{a_0}{2} \left[3n^2-l\left(l+1\right) \right], $$

de manera que


$$ \Delta x \approx \Delta z \approx \frac{1}{\sqrt{3}}R_{nl}.$$

Por otra parte, en el límite $\gamma\rightarrow\infty$, Potekhin (1998) presenta la siguiente aproximación:

$$\Delta x \approx a_0 \sqrt{\frac{|m|+1}{\gamma}},$$

y

$$\Delta z \approx \left [ \frac{a_0}{{\sqrt{2}}}+\frac{a_0}{\ln\left[\gamma/\left(1+|m|\right)\right]} \right ]  \left(\frac{Ry}{E_{m\nu}\left(\gamma\right)}\right)^{1/2},\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\left(\nu=0\right)$$

$$\Delta z \approx 1.6a_0 \frac{Ry}{E_{m\nu}\left(\gamma\right) },\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\left(\nu>0\right)$$

donde $E_{m\nu}$ es la energía del estado expresada en $Ry$. Ya tengo escrito un algoritmo que ajusta estas energías. Se puede observar que el ajuste de Potekhin es bastante bueno para la componente transversal, $\Delta x$, pero falla a la hora de ajustar $\Delta z$ (sobre todo a $m$ y $\nu$ grandes). Esta discordancia en $\Delta z$ podría ser mejorada si se hace

$$ \Delta z \approx a_0 \alpha \frac{Ry}{E_{m\nu}\left(\gamma\right)}, $$

con $\alpha$ una constante arbitraria.

En cualquier caso, se me ocurre que podría trabajar con funciones sigmoid en los planos $\log_{\gamma}-\Delta x$ y $\log_{\gamma}-\Delta z$ estableciendo algunas condiciones:

- En el límite $\gamma\rightarrow 0$ la función debe tender a 1.
- La función debe variar suavemente, con lo cual puedo descartar puntos que produzcan un cambio abrupto de la pendiente.
- La función debe presentar convergencia con $m\rightarrow -\infty$.
